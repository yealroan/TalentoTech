```python
import pandas as pd
```

```python
df1 = pd.read_excel("/content/example_ventas1.xlsx")
```

```python
print("Ejercicio1")
print (df1)
```

```python
ventas_david = df1[df1['Vendedor']== 'Sara']
print (ventas_david)
```

```python
print("Ejercicio2 calcular el total de ventas por region")
ventas_por_region = df1.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
print("df1")
```

```python
#Ejercicio 4 calcular el promedio de ventas por region
print ("Ejercicio 4 calcular el promedio de ventas por region")
ventas_por_producto = df1.groupby('Region')['Ventas'].mean()
print(ventas_por_producto)
```

```python
#Ejercicio 5 ordenar el df por las ventas en orden descendentes
print("Ejercicio 5 ordenar el dataframe por ventas en orden descendente")
df1_ordenado = df1.sort_values  (by='Ventas',ascending=False)
print(df1_ordenado)
```

```python
#Ejercicio 6 filtrar las filas con ventas mayores a 10.000
print("Ejercicio 6 filtrar las filas con ventas mayores a 10.000")
df1_filtrado = df1[df1['Ventas']>10000]
print (df1_filtrado)
```

```python
#Ejercicio 7 filtrar las filas con ventas > 10000 y ordenar de mayor a menor
print("Ejercicio 7: filtrar las filas con ventas mayores a 10000 y ordenar de mayor a menor")
df1_filtrado = df1[df1['Ventas']>10000].sort_values(by='Ventas', ascending=True)
print(df1_filtrado)
```

```python
# Ejercicio 9 eliminar la columna vendedor
print('Ejercicio 8 Eliminar la columna vendeor')
df1_sin_vendedor = df1.drop(columns=['Vendedor'])
print(df1_sin_vendedor)

```
